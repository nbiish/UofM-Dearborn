#### Categories of Economics
- Factors of production: Factors of production are the inputs used to produce goods and services. The most basic factors are land, labor, capital, and entrepreneurship.
    - Land: Land includes not only the ground but also any natural resources on it, such as oil, gold, or water.
    - Capital: Capital includes the buildings, machines, and tools used in the production process.
        - Things that nature did **NOT** provide but that are used to produce goods and services.
        - Technology is a type of capital.
        - Nothing to do with money, stock, or bonds in microeconomics.
    - Labor: Labor is the effort that people contribute to the production of goods and services.
        - The physical and mental effort of people.
        - The quality of labor depends on the skills, education, and motivation of the workers.
    - Entrepreneurship: Entrepreneurship is the effort used to coordinate the other factors of production to produce and sell products.
        - The entrepreneur is the person who takes the risk of starting a new business.
        - The entrepreneur is the person who makes the decisions about what and how to produce.
        - The entrepreneur is the person who innovates.

### Production Possibility Frontier (PPF) Graph Understanding

```python
# Quadrant 1 Graph
# x axis = Capital Goods
# y axis = Consumer Goods
# Curve declines from left to right

# Create
# Create new data for PPF curve (concave curve)
x_ppf = np.linspace(0, 10, 100)
y_ppf = 10 * (1 - (x_ppf/10)**2)**0.5  # Creates a concave curve

plt.figure(figsize=(10, 6))
plt.plot(x_ppf, y_ppf, 'b-', label='PPF')
plt.title('Production Possibility Frontier (PPF)')
plt.xlabel('Capital Goods')
plt.ylabel('Consumer Goods')
plt.grid(True)
plt.legend()
plt.axis([0, 10, 0, 10])  # Set axis limits
plt.show()
```

## Chapter 3 - Demand, Supply, and Market Equilibrium

### Demand Curves:
- The relationship between the price (p) and quantity (q) presented graphically is a demand curve.
- Demand curves has a downward or negative slope.
    - indicating that lower prices lead to higher quantities demanded.

<u>**Definition**</u>:
- Demand refers to the number of units of a good or service <u>consumers are willing and able to purchase</u> at various prices over a given period of time.

- **KEY POINTS OF DEFINITION**:
    - **Willing**: Consumers must be willing to buy the good or service.
    - **Able**: Consumers must have the resources to buy the good or service.
    - **Purchase**: Consumers must actually buy the good or service.
    - **Various Prices**: Demand is not just the quantity of a good or service that consumers are willing and able to buy at one price.
    - **Given Period of Time**: Demand is not just the quantity of a good or service that consumers are willing and able to buy at one point in time.

### The Law of Demand:
- Demand: The relationship between the price of a good and the quantity demanded.
    - Notation: Px

- Quantity demanded: The amount of a good that consumers are willing and able to purchase at a given price.
    - Notation: QDx

<u>**Definition**</u>:
- There exists a negative or inverse relationship between the price of Px (demand) and QDx (quantity demanded), ceteris paribus (all your other things are equal).

- **KEY POINTS OF DEFINITION**:
    - **Negative Relationship**: As the price of a good increases, the quantity demanded decreases.
    - **Inverse Relationship**: As the price of a good increases, the quantity demanded decreases.
    - **Ceteris Paribus**: All other things are equal.
        - The law of demand holds true only if all other factors that affect demand are held constant.

### Curve Mathematical Definition:
- delta(m) Px --> delta(m) QDx
- decrease in Px --> increase in QDx

```python
# Quadrant 1 Graph
# x axis =  QDx
# y axis = Px
# Px = endogenous variable
# QDx = exogenous variable
# Curve declines from left to right with convex curve

# Create
# Create new data for demand curve (convex curve)
x_demand = np.linspace(0, 10, 100)
y_demand = 10 - x_demand  # Made slope steeper by changing from 0.5 to 1

plt.figure(figsize=(10, 6))
plt.plot(x_demand, y_demand, 'b-', label='Demand Curve')
plt.title('Demand Curve')
plt.xlabel('Quantity Demanded')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.axis([0, 10, 0, 10])  # Set axis limits
plt.show()


```
