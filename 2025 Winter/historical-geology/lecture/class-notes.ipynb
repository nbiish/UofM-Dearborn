 ## Earth Materials - Minerals and Rocks

 ### Minerals:

- geological definition of a mineral:
    - naturally occuring, inorganic
    - crystalline solid
        - crystalline: atoms arranged in specific 3d framework
    - minerals have a narrowly defined chemical composition
    - characteristic physical properties
        - density
        - hardness
        - color

- Minerals - The building blocks of rocks
    - a minerals composition is shown by a chemical formula
        - a shorthand way of indicating how many atoms of differnet kinds it contains
        - quartz molecules consist of 1 silicon atom and 2 oxygen atoms
            - Quartz: SiO2
            - Ratio of silicon to oxygen is 1:2
        - Orthoclase

- Native elements:
    - a few minerals consist of only one element
        - gold, silver, copper, sulfur, diamond
        - they are not compounds

    - Gold: Au
    - Diamond: C
    
- Mineral Properties:
    - controlled by:
        - chemical composition
        - crystalline structure
    
    - Useful for identification:
        - color
        - streak
        - luster
        - crystal form
        - cleavage
            - tendency to break along planes of weak bonding
            - different cleavage planes are not parallel 
        - fracture
        - hardness
            - 1-10 scale
            - talc is 1
            - diamond is 10
        - specific gravity

- More than 3500 minerals are known
    - only about 2 dozen are common
    - many are important economically
        - gold, silver, copper, aluminum, iron, lead, zinc, mercury, sulfur, diamond, quartz, feldspar, mica, clay minerals
    - mineral groups
        - minerals with similar chemical compositions and structures
        - minerals with similar negatively charged ions or ion groups

- Silicates:
    - Si and O
    - make up perhaps 95% of the earth's crust
    - account for about 1/3 of all known minerals

    - the basic building blocks of silicates:
        - silica tetrahedron
            - 1 silicon atom surrounded by 4 oxygen atoms
            - SiO4

        - ferromanagnesian silicates (MAFIC) 
            - contains iron (Fe)
            - contains magnesium (Mg)

        - nonferromagnesian silicates (FELSIC)
            - contains potassium (K)
            - contains sodium (Na)
            - contains calcium (Ca)
            - contains aluminum (Al)

- Silica tetrahedra can be:
    - isolated
    - linked in chains
    - complex 3d networks

- Ferromagnesian silicates:
    - olivine
    - augite
    - hornblende
    - biotite

- Nonferromagnesian silicates:
    - orthoclase
    - muscovite
    - quartz
    - plagoclase

- other mineral groups:
    - carbonates
        - minerals with carbonate ion (CO3)2-
        - calcite CaCO3
            - constituent of limestone
        - dolomite CaMg(CO3)2
            - constituent of dolostone

- Rock forming minerals;
    - one or more minerals
    - hundreds of minerals occur in rocks
    - only a few are common
        - called rock forming minerals
    - most rock-forming minerals are silicates
        - but carboaates, sulfates, and halides are also common

- rock cycle
    - events that lead to origin, destruction, and reformation of rocks

- igneous rocks
    - form from solidification of molten rock
    - molten rock is called magma
    - if magma solidifies beneath the surface, it forms intrusive igneous rocks
        - granite
    - if magma solidifies at the surface, it forms extrusive igneous rocks
        - basalt

    - magma is moleten material
        - below the surface

    - lava is molten material
        - at the surface

    - pyroclatic materials
        - are particles less than 2mm in diameter


- categories of igneous rocks:
    - extrusive or volcanic rocks
        - formed at the surface
        - from lava or pryoclastic materials
    - intrusive or plutonic rocks
        - formed from magma ingected into the crust
        - or formed in place in the crust
        - **plutons** are intrusive bodies
            - dike
                - horizontal intrusion
            - sill
                - vertical intrusion
                - branch off dikes
            - volcanic pipe expels to the surface
            
- igneous rock textures:
    - texture:
        - size, shape, and arrangement of mineral grains
    - texture is controlled by:
        - rate of cooling
        - amount of gas in the magma
        - presence of water in the magma
        - presence of other minerals in the magma

    - cooling-rate textures:
        - phaneritic
            - coarse grained
            - slow cooling
        - aphanitic
            - fine grained
            - rapid cooling
        - porphyritic
            - large crystals in a fine grained matrix
            - two stages of cooling
            - mixture of large and small crystals
            - multiple textures
            - phenocrysts are large crystals surrounded by by a fine grained groundmass
                - cooled slowly intrusively and expelled to surface
        - glassy
            - with no grains
                - cooled too quickly for minerals to grow

    - vesicular
        - with holes
        - gas bubbles in the rock
        - gas bubbles are called vesicles
        - vesicles are formed by gas escaping from the magma
        - pumice and scoria are vesicular rocks
            - pumice is light and floats
                - used for exfoliating skin
                - egypitians used it for cleaning and shaving
                - microscopically porous glass
            - scoria is heavy and sinks
                - used for landscaping
                - vesicular basalt
        - rapid-cooling
            - glassy texture
            - no crystals
            - obsidian is a glassy rock
                - used for making arrowheads
                - used for making surgical instruments
                - used for making jewelry
        - slow-cooling
            - phaneritic texture
            - large crystals
            - granite is a phaneritic rock
                - used for countertops
                - used for monuments
                - used for buildings
                - used for sculptures

- classifying igneous rocks:
    - based on mineral composition
        - light colored rocks are nonferromagnesian minerals
        - imtermediate colored rocks are a mix of felsic and mafic minerals
        - dark colored rocks are ferromagnesian minerals
    - based on texture

- common igneous rocks:
    - mafic rocks:
        - basalt
        - gabbro

    - intermediate rocks:
        - andesite
        - diorite
    
    - felsic rocks:
        - rhyolite
        - granite

- metamorphic rocks:
    - form from preexisting rocks
    - subjected to heat, pressure, and chemically active fluids
    - metamorphism
        - changes in mineralogy, texture, and chemical composition
        - <u>WITHOUT MELTING</u>

    - metamorphic rocks are classified by texture and mineral composition
        - compositional change
        - textural change

    - foliated rocks
        - have a planar fabric
        - slate
        - phyllite
        - schist
        - gneiss

    - nonfoliated rocks
        - do not have a planar fabric
        - marble
        - quartzite
        - hornfels
        - anthracite

    - agents of metamorphism:
        - heat
            - increases rate of chemical reactions
            - recrystallization of minerals
        - pressure
            - lithostatic pressure
                - pressure from all directions
            - differential pressure
                - pressure from one direction

        - chemically active fluids ESPECIALLY water
            - water is the most important agent
            - water is a solvent
            - water is a transport medium
            - water is a catalyst
            - water is a source of ions

        - contact metamorphism
            - rocks are heated by contact with magma
            - rocks are altered by heat and chemically active fluids
            - rocks are altered by pressure
            - rocks are altered by differential stress
            - rocks are altered by recrystallization
            - rocks are altered by metasomatism
                - the process of changing a rock's chemical composition by adding or removing elements

        - regional metamorphism
            - most common type of metamorphism
            - rocks are subjected to heat and pressure
            - rocks are subjected to differential stress
            - rocks are subjected to chemically active fluids
            - rocks are subjected to recrystallization
            - rocks are subjected to metasomatism
            - occurs at convergent plate boundaries

- bedding:
    - sedimentary rocks typically exhibit layers or stratification
    - individual layers less than 1cm thick are called laminations
