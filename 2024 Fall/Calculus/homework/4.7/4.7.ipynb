## Problem 1

```python
# Problem:
# If you have 40 meters of fencing and want to enclose a rectangular area up against a long, straight wall, what is the largest area you can enclose?

# Step 1: Define the problem.
# The problem is asking for the largest area that can be enclosed by a rectangular fence with 40 meters of fencing.

# Step 2: Visualize the problem.
# The rectangular fence will have a length and width. The length will be up against the wall, and the width will be perpendicular to the wall.

# Step 3: Define the variables.
# Let x be the width of the rectangular fence.
# Let y be the length of the rectangular fence.
# DESMOS INPUT: x = 10, y = 20

# Step 4: Write the equation for the area of the rectangle.
# The area of a rectangle is given by A = x*y.
# DESMOS INPUT: A = 10*20

# Step 5: Write the equation for the perimeter of the rectangle.
# The perimeter of a rectangle is given by P = 2x + y = 40.
# DESMOS INPUT: 2x + y = 40

# Step 6: Solve the perimeter equation for y.
# y = 40 - 2x.
# DESMOS INPUT: y = 40 - 2x

# Step 7: Substitute the expression for y into the area equation.
# A = x*(40 - 2x).
# DESMOS INPUT: A = x*(40 - 2x)

# Step 8: Expand the area equation.
# A = 40x - 2x^2.
# DESMOS INPUT: A = 40x - 2x^2

# Step 9: Find the critical points of the area function.
# The critical points occur where the derivative of the area function is zero or undefined.
# DESMOS INPUT: dA/dx = 40 - 4x

# Step 10: Find the derivative of the area function.
# dA/dx = 40 - 4x.
# DESMOS INPUT: dA/dx = 40 - 4x

# Step 11: Set the derivative equal to zero and solve for x.
# 40 - 4x = 0.
# 4x = 40.
# x = 10.
# DESMOS INPUT: x = 10

# Step 12: Find the second derivative of the area function.
# d^2A/dx^2 = -4.
# DESMOS INPUT: d^2A/dx^2 = -4

# Step 13: Determine the nature of the critical point.
# The second derivative is negative, so the critical point is a maximum.
# DESMOS INPUT: Second derivative test

# Step 14: Find the maximum area.
# Substitute x = 10 back into the area equation.
# A = 40*10 - 2*10^2.
# A = 400 - 200.
# A = 200.

# Step 15: Answer the question.
# The largest area that can be enclosed by a rectangular fence ...
# ... with 40 meters of fencing is 200 square meters.

# VISUALIZATION IN NEXT CODE CELL
```

```python
# Step 16: Graph the area function.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the area function
def area(x):
    return 40*x - 2*x**2
```

```python
# Generate x values
x = np.linspace(0, 20, 100)
y = area(x)
```

```python
# Plot the area function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Area Function: A = 40x - 2x^2')
plt.xlabel('Width (x)')
plt.ylabel('Area (A)')
plt.title('Area of Rectangular Fence vs. Width')
plt.axvline(x=10, color='r', linestyle='--', label='Maximum Area')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Display step 13's meaning with matplotlib
x = np.linspace(0, 20, 100)
y = 40 - 4*x

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='dA/dx = 40 - 4x')

plt.axhline(y=0, color='r', linestyle='--', label='Critical Point')
plt.xlabel('Width (x)')
plt.ylabel('dA/dx')
plt.title('Derivative of Area Function vs. Width')

plt.legend()
plt.grid(True)
plt.show()

```
